```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

N = 7
initCondi = [0.4, 0.25, 0.3]
r = 0.025
d = 0.145

wl = np.linspace(1, 1, N)
wr = np.linspace(1, 1, N)
t = np.linspace(0, 3, N)

phi_t = [initCondi[0]]
for i in range(len(t)-1):
    calka = phi_t[i] + np.trapz([(wr[i]-wl[i])*r/d, (wr[i+1]-wl[i+1])*r/d],[t[i], t[i+1]])
    phi_t.append(calka)
    
x_t = [initCondi[1]]
for i in range(len(t)-1):
    calka = x_t[i] + np.trapz([(wr[i]+wl[i])*r/2*np.cos(phi_t[i]), (wr[i+1]+wl[i+1])*r/2*np.cos(phi_t[i+1])],[t[i], t[i+1]])
    x_t.append(calka)

y_t = [initCondi[2]]
for i in range(len(t)-1):
    calka = y_t[i] + np.trapz([(wr[i]+wl[i])*r/2*np.cos(phi_t[i]), (wr[i+1]+wl[i+1])*r/2*np.cos(phi_t[i+1])],[t[i], t[i+1]])
    y_t.append(calka)

plt.plot(x_t, y_t, label="środek robota")
plt.xlabel("oś X, N = 7, t = 3")
plt.ylabel("oś Y")
plt.legend(loc=4)
plt.savefig("2cMaleN.pdf", format='pdf')
```
