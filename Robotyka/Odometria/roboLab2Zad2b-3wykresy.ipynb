```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

N = 50
initCondi = [0, 0, 0]

t = np.linspace(0, 3, N)

phi_t = [initCondi[0]]
for i in range(len(t)-1):
    calka = phi_t[i] + np.trapz([np.cos(t[i]), np.cos(t[i+1])],[t[i], t[i+1]])
    phi_t.append(calka)
    
x_t = [initCondi[1]]
for i in range(len(t)-1):
    calka = x_t[i] + np.trapz([np.sin(t[i])*np.cos(phi_t[i]), np.sin(t[i+1])*np.cos(phi_t[i+1])],[t[i], t[i+1]])
    x_t.append(calka)

y_t = [initCondi[2]]
for i in range(len(t)-1):
    calka = y_t[i] + np.trapz([np.sin(t[i])*np.sin(phi_t[i]), np.sin(t[i+1])*np.sin(phi_t[i+1])],[t[i], t[i+1]])
    y_t.append(calka)

plt.plot(t, phi_t, label="phi(t)")
plt.plot(t, x_t, label="x(t)")
plt.plot(t, y_t, label="y(t)")
plt.xlabel("t = 3")
plt.ylabel("wartość")
plt.legend(loc=4)
plt.savefig("2bOdCzasu.pdf", format='pdf')
```
